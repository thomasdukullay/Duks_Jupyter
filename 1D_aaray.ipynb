```python
ar = []
print('How many elements to store inside the array?', end="")
num = input()
print("\nEnter", num, "elements:", end="")
num = int(num)
for i in range(num):
  element = input(f"Enter element {i+1}: ")
  ar.append(element)

print("\nThe array elements are:")
for i in range(num):
  print(ar[i], end=" ")
```
