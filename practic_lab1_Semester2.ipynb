```
n = int(input("Enter the number of element"))
arr = []
for i in range(n):
  num= int(input(f"Enter element{i+1}:"))
  arr.append(num)

total= 0
for num in arr:
  total += num
print("\nsum of all element =", total)
```

```
class stack:
  def __init__(self, size):
    self.stack = []
    self.size = size
  def isEmty(self):
    return len(self.stack) == 0
  def isFull(self):
    return len(self.stack) == self.size
  def push(self, item):
    if self.isFull():
      print("Stack is full")
    else:
      self.stack.append(item)
      print(item, "pushed into stack")
  def pop (self):
    if self.isEmty():
      print("Stack is empty")
    else:
      print(self.stack.pop(), "poped from stack")
  def peek(self):
    if self.isEmty():
      print("Stack is empty")
    else:
      print("Top element is", self.stack[-1])

# The interactive part of the code should not be part of the class definition.

print("\nstack operation")
stack_size = int(input("Enter stack size: "))
stack = stack(stack_size)
while True:
  print("1. Push\n2. Pop\n3. Peek\n4. Exit\n5. Is Empty\n6. Is Full")
  choice = int(input("Enter your choice: "))
  if choice == 1:
    item = int(input("Enter element to push: "))
    stack.push(item)
  elif choice == 2:
    stack.pop()
  elif choice == 3:
    stack.peek()
  elif choice == 4:
    print("Exiting program.")
    break
  elif choice == 5:
    print ("Stack Empty: ", stack.isEmty())
  elif choice == 6:
    print ("Stack Full: ", stack.isFull())
  else:
    print("Invalid choice")
```
