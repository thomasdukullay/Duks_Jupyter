```
print("Enter the size of array", end=": ")
tot = int(input())
arr = []
print(f"Enter {tot} elements:")

for i in range(tot):
    arr.append(input(f"Element {i+1}: "))

print("\nEnter the value to delete:", end=" ")
Val = input()

if Val in arr:
    arr.remove(Val)
    print("\nThe new array is:")
    for item in arr:
        print(item, end=" ")
    print() # Add a newline after printing the array
else:
    print("\nElement doesn't exist in the array")
```
