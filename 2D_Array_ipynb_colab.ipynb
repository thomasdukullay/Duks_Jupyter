```
r_num = int(input("Enter the number of rows: "))
c_num = int(input("Enter the number of columns: "))

# Initialize a 2D list (matrix) with zeros
trend_arr = [[0 for _ in range(c_num)] for _ in range(r_num)]

for row in range(r_num):
    for col in range(c_num):
        trend_arr[row][col] = row * col

print(trend_arr)
```
