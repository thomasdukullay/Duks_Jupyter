```python
class Node:
    def __init__(self, data):
        self.data = data  # Assign data from the parameter
        self.next = None  # Initialize next as null (no next node yet)

class LinkedList:
    def __init__(self):
        self.head = None  # Initialize head of the list as None

    def append(self, data):
        new_node = Node(data)  # Use the data parameter to create the new node
        if self.head is None:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node

    def display(self):
        current = self.head
        elements = []
        while current:
            elements.append(current.data)
            current = current.next
        print(" -> ".join(map(str, elements)))
```

### Example Usage

```python
class Node:
    def __init__(self, data):
        self.data = data  # Assign data from the parameter
        self.next = None  # Initialize next as None

class LinkedList:
    def __init__(self):
        self.head = None  # Initialize head of the list as None

    def append(self, data):
        new_node = Node(data)  # Use the data parameter to create the new node
        if self.head is None:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node

    def display(self):
        current = self.head
        elements = []
        while current:
            elements.append(current.data)
            current = current.next
        print(" -> ".join(map(str, elements)))


my_list = LinkedList()

# Add elements to the list
my_list.append(1)
my_list.append(2)
my_list.append(3)
my_list.append(4)
my_list.append(5)
my_list.append(6)
my_list.append(7)

# Display the linked list
print("Linked List:")
my_list.display()

# Add another element
my_list.append(5)
print("Linked List after appending 5:")
my_list.display()

my_list.append(6)
print("Linked List after appending 6:")
my_list.display()

my_list.append(7)
print("Linked List after appending 7:")
my_list.display()
```
